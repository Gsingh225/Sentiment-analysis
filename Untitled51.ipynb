```python
import pandas as pd
import transformers
from transformers import GPT2Tokenizer
import torch
from torch.utils.data import Dataset, DataLoader
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("abhi8923shriv/sentiment-analysis-dataset")

print("Path to dataset files:", path)
```

```python
#read dataset

train_data_path = path + '/train.csv'
test_data_path = path + '/test.csv'

train_data = pd.read_csv(train_data_path, encoding='latin1')
test_data = pd.read_csv(test_data_path, encoding='latin1')

print(train_data.head().to_string())
print(test_data.head().to_string())
print('\n\n\n\n\n')
print(train_data.info())
print('\n\n\n\n\n')
print(test_data.head().to_string())

test_sentiment_counts = test_data['sentiment'].value_counts()

# Print the count of neutral sentiments in the test data
print(test_sentiment_counts['neutral'])
print(test_sentiment_counts['negative'])
print(test_sentiment_counts['positive'])
```

```python
train_data.drop(['textID', 'selected_text', 'Time of Tweet', 'Age of User', 'Country', 'Population -2020', 'Land Area (Km²)', 'Density (P/Km²)'], axis=1, inplace=True)
test_data.drop(['textID', 'Time of Tweet', 'Age of User', 'Country', 'Population -2020', 'Land Area (Km²)', 'Density (P/Km²)'], axis=1, inplace=True)

train_data.dropna(inplace=True)
test_data.dropna(inplace=True)

print(train_data.head().to_string())
```

```python
tokienizer = tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})

texts = train_data['text'].tolist()
tokenized_texts = tokenizer(texts, padding=True, truncation=True, return_tensors='pt')

test_texts = test_data['text'].tolist()
tokenized_test_texts = tokenizer(test_texts, padding=True, truncation=True, return_tensors='pt')

sents_to_num = {'neutral': 0, 'negative': 1, 'positive': 2}
train_data['sentiment_numeric'] = train_data['sentiment'].map(sents_to_num)
test_data['sentiment_numeric'] = test_data['sentiment'].map(sents_to_num)

train_labels = train_data['sentiment_numeric'].tolist()
test_labels = test_data['sentiment_numeric'].tolist()
```

```python
class SentimentDataset(Dataset):
  def __init__(self, encodings, labels):
    self.encodings = encodings #Tokenized inputs containing input_ids and attention_mask.
    self.labels = labels #corresponding sent labels

  def __len__(self):
    return len(self.labels)

  def __getitem__(self, idx):
    item = {key: val[idx] for key, val in self.encodings.items()}
    item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)
    return item
```

```python
train_dataset = SentimentDataset(tokenized_texts, train_labels)
test_dataset = SentimentDataset(tokenized_test_texts, test_labels)
```

```python
class MultiHeadAttention(nn.Module):
  def __init__(self, embed_dim, num_heads):
    super().__init__()
    self.embed_dim = embed_dim
    self.num_heads = num_heads
    self.head_dim = embed_dim // num_heads

    assert self.head_dim * num_heads == embed_dim, "Embed dim must be divisible by the number of heads"

    self.query = nn.Linear(embed_dim, embed_dim)
    self.key = nn.Linear(embed_dim, embed_dim)
    self.value = nn.Linear(embed_dim, embed_dim)
    self.fc_out = nn.Linear(embed_dim, embed_dim)

  def forward(self, x):
    batch_size, seq_len, embed_dim = x.shape
    Q = self.query(x).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1,2)
    K = self.key(x).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1,2)
    V = self.value(x).view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1,2)

    energy = torch.matmul(Q, K.transpose(-2, -1)) / (self.head_dim ** 0.5)
    attention = F.softmax(energy, dim=-1)
    out = torch.matmul(attention, V).transpose(1, 2).reshape(batch_size, seq_len, embed_dim)
    return self.fc_out(out)

class HybridCNNGRNAttention(nn.Module):
  def __init__(self, vocab_size, embed_dim, num_classes, num_heads=4, gru_hidden_size=128, cnn_out_channels=64):
    super().__init__()
    self.embedding= nn.Embedding(vocab_size, embed_dim)
    self.cnn = nn.Conv1d(embed_dim, cnn_out_channels, kernel_size=3, padding=1)
    self.gru = nn.GRU(cnn_out_channels, gru_hidden_size, batch_first=True, bidirectional=True)
    self.attention = MultiHeadAttention(gru_hidden_size * 2, num_heads)
    self.dropout = nn.Dropout(0.5)
    self.fc = nn.Linear(gru_hidden_size * 2, num_classes)

  def forward(self, input_ids):
        x = self.embedding(input_ids)  # (batch_size, seq_len, embed_dim)
        x = x.permute(0, 2, 1)  # (batch_size, embed_dim, seq_len) for CNN
        x = self.cnn(x)  # (batch_size, cnn_out_channels, seq_len)
        x = self.dropout(x) #drop
        x = x.permute(0, 2, 1)  # (batch_size, seq_len, cnn_out_channels) for GRU
        x, _ = self.gru(x)  # (batch_size, seq_len, gru_hidden_size * 2)
        x = self.dropout(x) #drop
        x = self.attention(x)  # (batch_size, seq_len, gru_hidden_size * 2)
        x = self.dropout(x) #drop
        x = x.mean(dim=1)  # Global average pooling
        return self.fc(x)
```

```python
vocab_size = len(tokenizer)
embed_dim = 768
num_classes = 3
num_heads = 8
gru_hidden_size = 256
cnn_out_channels = 64

model = HybridCNNGRNAttention(vocab_size, embed_dim, num_classes, num_heads, gru_hidden_size, cnn_out_channels)

batch_size = 32
shuffle_train = True

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=shuffle_train)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 5

train_losses = []
# Training loop
for epoch in range(epochs):
    model.train()  # Set the model to training mode
    running_loss = 0.0
    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}"):
        optimizer.zero_grad()  # Zero the gradients
        input_ids = batch['input_ids'].to(device)
        labels = batch['labels'].to(device)

        # Forward pass
        outputs = model(input_ids)
        loss = loss_fn(outputs, labels)

        # Backward pass and optimization
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    train_losses.append(epoch_loss)
    print(f"Epoch {epoch+1} Training Loss: {epoch_loss:.4f}")

# Plotting the training loss
plt.plot(train_losses)
plt.xlabel("Epoch")
plt.ylabel("Training Loss")
plt.title("Training Loss over Epochs")
plt.show()
```

```python
################################################################################
# 1) VALIDATION LOOP (using your existing test_loader)
################################################################################
model.eval()
val_loss = 0.0
val_correct = 0
with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids)
        loss = loss_fn(outputs, labels)
        val_loss += loss.item()

        preds = torch.argmax(outputs, dim=1)
        val_correct += (preds == labels).sum().item()

val_loss /= len(test_loader)
val_accuracy = val_correct / len(test_loader.dataset)
print(f"Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}")
```
